#### Problem 24
This problem is relatively straight forward. The left matrix

$$

\begin{bmatrix}
1 & a & b \\
0 & 1 & c \\
0 & 0 & 1 \\
\end{bmatrix}

$$

is $U$. As the problem states, take $U$, augment the identity matrix onto its right

$$

\begin{bmatrix}
1 & a & b & 1 & 0 & 0 \\
0 & 1 & c & 0 & 1 & 0 \\
0 & 0 & 1 & 0 & 0 & 1 \\
\end{bmatrix}

$$

and perform Gauss-Jordan elimination.

Side note: The Gauss part of elimination has been done (forward elimination) so the Jordan part of elimination is what remains.

For Gauss Jordan elimination here, we need to continue elimination on the left side until it is the identity matrix. The right side will then be $U^{-1}$. 

This leads to

$$
U^{-1} = 

\begin{bmatrix}
1 & -a & ac-b \\
0 &  1 &   -c \\
0 &  0 &    1 \\
\end{bmatrix}

$$
