#### Problem 28

Yes. If $x\cdot y < 0$, the angle between them is $\theta > 90°$. The problem is asking if there can be three vectors where the angle between each of them is > 90°. This is definitely possible: 3 vectors joined at their tails with more than 90 degrees between them can be drawn. An example would be slicing a circle 3 into three equal pieces, with the lines between the slices being the vectors. The angle between each vector is 360°/3 = 120°, which is greater than 90°.
