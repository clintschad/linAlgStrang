```python
import numpy as np

A = np.array([[0.8,0.3],
              [0.2,0.7]])
u1 = np.array([0.6,0.4])

Au1 = np.dot(A,u1)

print("Matrix A = ")
print(A)
print("\n")

print("u1 = ")
print(u1)
print("\n")

print("Au1 = ")
print(Au1)
```

Used code above to keep multiplying A by consecutive u's and generate more u's to find a pattern. After generating 5 u's, this is what I found:

* The components of every u add to 1.
* To find the next u, 
    * The first element in u: take the difference between current and previous. Divide by 2. Subtract from current.
    * The second element in u: take the difference between current and previous. Divide by 2. Add to current.
