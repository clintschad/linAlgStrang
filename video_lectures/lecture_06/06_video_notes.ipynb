### Vector Space

A vector space is a space (same as group?) where
* adding any two vectors in the space also produces a vector in the space
* multiplying any vector in the space by a scalar produces a vector in the space

In summary, any linear combination of vectors in the space produce a vector still in the space.


### Subspaces
A subspace is a vector space inside another vector space.

For example, a plane (going through the origin) in 3D space is a subspace: linear combinations of vectors in the plane stay in the plane. The plane is a subspace of $\mathbb{R}^{3}$.

Another example is a line going through the origin in 3D space. Linear combos of vectors on the line stay on the line. The line is a subspace of $\mathbb{R}^{3}$.

Subspaces must contain the origin (the zero vector). For example, a plane or line in $\mathbb{R}^{3}$ not going through the origin is not a subspace of $\mathbb{R}^{3}$.

The intersection of subspaces, e.g. $A \cap B$, is a subspace.

The union subspaces, e.g. $A \cup B$, is not a subspace.

### Column Space

Example

$$
A =
\begin{bmatrix}
1 & 1 & 2 \\
2 & 1 & 3 \\
3 & 1 & 4 \\
4 & 1 & 5 \\
\end{bmatrix}
$$

Column space of $A$ is a subspace of $\mathbb{R}^{4}$.

All linear combinations of the columns of $A$.

Does $Ax = b$ have a solution for every $b$?

$$
A =
\begin{bmatrix}
1 & 1 & 2 \\
2 & 1 & 3 \\
3 & 1 & 4 \\
4 & 1 & 5 \\
\end{bmatrix}
\begin{bmatrix}
x_{1} \\
x_{2} \\
x_{3} \\
\end{bmatrix}
=
\begin{bmatrix}
b
\end{bmatrix}
$$

Which $b's$ allow the system to be solved? --> Can solve $Ax = b$ exactly when $b$ is in $C(A)$, i.e. the column space of $A$.

All three columns are not independent. Column 3 is the sum of the first two columns and is not contributing anything. It can technically be "thrown" out and the same column space will still remain.

The columns that do contribute to constructing the column space are called _pivot columns_.

The column space of $A$ is a 2D subspace (plane) of $\mathbb{R}^{4}$.


### Null Space

Keeping with the previous example, the null space of $A$ contains all $x = [ x_{1} x_{2} x_{3}]$ that make $Ax = 0$.

Since $x$ has 3 components, the null space is in $\mathbb{R}^{3}$.

Null space $N(A)$ always contains the 0 vector.

Example

$$
Ax = 
\begin{bmatrix}
1 & 1 & 2 \\
2 & 1 & 3 \\
3 & 1 & 4 \\
4 & 1 & 5 \\
\end{bmatrix}
\begin{bmatrix}
x_{1} \\
x_{2} \\
x_{3} \\
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0 \\
0 \\
0 \\
\end{bmatrix}

$$

The null space are the $x's$ that give 0.

Below is an $x$ that gives 0.

$$
Ax = 
\begin{bmatrix}
1 & 1 & 2 \\
2 & 1 & 3 \\
3 & 1 & 4 \\
4 & 1 & 5 \\
\end{bmatrix}
\begin{bmatrix}
1  \\
1  \\
-1 \\
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0 \\
0 \\
0 \\
\end{bmatrix}

$$

All the $x's$ -- the null space -- that make $Ax = 0$ are the vectors $c[1, 1, -1]$. See below.

$$
Ax = 
\begin{bmatrix}
1 & 1 & 2 \\
2 & 1 & 3 \\
3 & 1 & 4 \\
4 & 1 & 5 \\
\end{bmatrix}
c
\begin{bmatrix}
1  \\
1  \\
-1 \\
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0 \\
0 \\
0 \\
\end{bmatrix}

$$

The null space here is a line in $\mathbb{R}^3$ through the origin.

How do we know this is a _space_? _Space_ indicates that the requirements for a space are satisfied.

Need to check that solutions to $Ax = 0$ always give a subspace.

If $Av = 0$ and $Aw = 0$, then $A(v+w) = 0$.

We also know this, because since addition is distributive, $A(v+w) = Av+Aw = 0 + 0 = 0$.

### Non Subspace Example

$$
Ax = 
\begin{bmatrix}
1 & 1 & 2 \\
2 & 1 & 3 \\
3 & 1 & 4 \\
4 & 1 & 5 \\
\end{bmatrix}
\begin{bmatrix}
x_{1} \\
x_{2} \\
x_{3} \\
\end{bmatrix}
=
\begin{bmatrix}
1 \\
2 \\
3 \\
4 \\
\end{bmatrix}

$$

While there are $x's$ that satisfy this matrix equation, such as $[1,0,0]$ and $[0,-1,1]$, the $x's$ don't form a subspace because the 0 vector (or the origin) isn't included.

This is like a line or plane that doesn't go through the origin.
