## Elimination for a Rectangular Matrix
Performing elimination on the matrix below

$$
A = 
\begin{bmatrix}
1 & 2 & 2 &  2 \\
2 & 4 & 6 &  8 \\
3 & 6 & 8 & 10 \\
\end{bmatrix}
$$

results in

$$
U = 
\begin{bmatrix}
1 & 2 & 2 & 2 \\
0 & 0 & 2 & 4 \\
0 & 0 & 0 & 0 \\
\end{bmatrix}
$$

This is in _echelon_ form, meaning it's not a nice upper triangle, but more of a staircase.

The number of pivots is 2, the first non-zero in the row, moving from top to bottom row.

The rank of the matrix is the number of pivots.

The columns with the pivots are called _pivot columns_.

The other columns not containing pivots are called _free columns_.

The solution to $Ux = 0$ is $x$, where $x = \begin{bmatrix}x_{1} & x_{2} & x_{3} & x_{4} \end{bmatrix}$

Since $U$ has _free columns_ in columns 2 and 4, we are _free_ to choose $x_{2}$ and $x_{4}$.

Suppose the following solution vector, $x = \begin{bmatrix}x_{1} & 1 & x_{3} & 0 \end{bmatrix}$ where $x_{2} = 1$ and $x_{4} = 0$, is chosen.

$Ux = 0$ in equation form is

$$
x_{1} + 2x_{2} + 2x_{3} + 2x_{4} = 0\\
2x_{3} + 4x_{4} = 0
$$

Substituting in our chosen $x_{2} = 1$ and $x_{4} = 0$ and doing back substitution, the _special_ solution vector is now $x = \begin{bmatrix} -2 & 1 & 0 & 0 \end{bmatrix}$.

$x$ is a vector in the null space since it's a solution to $Ux = 0$.

Any scalar multiple of this vector is in the null space, so this vector defines a line in the null space, as $cx$.

This line, however, doesn't encompass the entire null space. There are 2 free variables, so there should be 2 independent vectors creating the null space, i.e. the null space should be a plane.

Since $x_{2}$ and $x_{4}$ were chosen to be 1 and 0 respectively, we will now choose them to be 0 and 1. This choice of 0 and 1 for both vectors is to ensure the _special_ solution vectors are independent of each other.

With $x_{2} = 0$ and $x_{4} = 1$ we repeat what we did before, by back substituting these into the system of equations and solving for the rest of the variables $x_{1}$ and $x_{3}$. This produces $x = \begin{bmatrix} 2 & 0 & -2 & 1 \end{bmatrix}$. As with the first special solution vector, this one also can be multiplied by any scalar, so the line in the null space that this represents is $dx$, where $d$ is a scalar.

These two vectors are _special_ solutions. The entire null space is the sum of the special solutions, i.e. the two vectors (lines) we found

$$
c
\begin{bmatrix}
-2 \\
 1 \\
 0 \\
 0
\end{bmatrix}

+

d
\begin{bmatrix}
 2 \\
 0 \\
-2 \\
 1
\end{bmatrix}

$$

The null space contains exactly all the combinations of the special solutions.

The number of special solutions is the number of free variables, i.e. the rank of the matrix.

For an _m x n_ matrix with rank $r$, the number of free variables is $n - r$.

This null space are all the solutions $x$ that satisfy $Ax = 0$.

## Reduced Row Echelon Form

The "upper triangular" matrix $U$ above is in _echelon_ form.

This matrix can be further reduced where elimination is continued to zero out as many non-pivot entries as possible and making the pivots equal 1.

Starting with $U$

$$
U = 
\begin{bmatrix}
1 & 2 & 2 & 2 \\
0 & 0 & 2 & 4 \\
0 & 0 & 0 & 0 \\
\end{bmatrix}
$$

Subtract row 2 from row 1.

$$
\begin{bmatrix}
1 & 2 & 0 & -2 \\
0 & 0 & 2 &  4 \\
0 & 0 & 0 &  0 \\
\end{bmatrix}
$$

First pivot is already 1. However, the second pivot is 2. Make 1 by dividing the second row by 2.

$$
R =
\begin{bmatrix}
1 & 2 & 0 & -2 \\
0 & 0 & 1 &  2 \\
0 & 0 & 0 &  0 \\
\end{bmatrix}
$$

This matrix $R$ is in the _reduced row echelon form_ of the original matrix $A$ from the start of this section. RREF is a common command used to compute this, such as when using [MATLAB](https://www.mathworks.com/help/matlab/ref/rref.html) or [Sympy](https://docs.sympy.org/latest/tutorials/intro-tutorial/matrices.html).

A matrix in RREF gives the pivot rows and pivot columns.

The pivots have 0's above and below them.

If the pivot columns (columns 1 and 3) are extracted to form a matrix, the matrix is an identity matrix $I$.

The free variable columns (columns 2 and 4) can also be extracted to form a matrix, when can be called $F$.

The null space matrix $N$ can be formed from $I$ and $F$, such that $AN = 0$, where $N$ are the columns of the special solutions. $N$ is constructed as

$$
N = 
\begin{bmatrix}
-F \\
I  \\
\end{bmatrix}
$$

$I$ is columns 1 and 2 of $R$ (not including the zeros from the zero rows)

$$
\begin{bmatrix}
1 & 0 \\
0 & 1 \\
\end{bmatrix}
$$

$F$ is columns 2 and 4 of $R$ (not including the zeros from the zero rows)

$$
\begin{bmatrix}
2 & -2 \\
0 &  2 \\
\end{bmatrix}
$$

Using the definition of $N$ above to construct is

$$
N = 
\begin{bmatrix}
-F \\
I  \\
\end{bmatrix}

=
\begin{bmatrix}
-2 &  2 \\
 1 &  0 \\
 0 & -2 \\
 0 &  1 \\
\end{bmatrix}
$$

Here, $F$ and $I$ don't stack on top of each other nicely. Since $I$ was _columns_ 1 and 3 of $R$, it becomes _rows_ 2 and 4 of $N$. Similarly, since $F$ was _columns_ 2 and 4 of $R$, $-F$ becomes _rows_ 1 and 2 of $N$.

Observe how these columns of $N$ are the same as the special solution vectors found at the end of the previous section.

The solutions to $U$ and $R$ are the same.
