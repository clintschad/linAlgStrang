#### Chapter 2

##### Matrix Transpose

The transpose of a symmetric matrix is the same matrix, such that $A^{T} = A$.

The transpose of a matrix times the original matrix results in a symmetric matrix. In other words, $A^{T}A$ is symmetric.

#### Chapter 3

#### Vector Spaces

A space in which vectors inside if linearly combined (scaled and added) always produce another vector inside the space. Vectors in the space cannot be combined to produce a vector outside the space.

##### Good Examples
$R^{2}$: all 2-dimensional real vectors, such as

\begin{bmatrix}
3 \\
2 \\
\end{bmatrix}

\begin{bmatrix}
0 \\
0 \\
\end{bmatrix}

\begin{bmatrix}
\pi \\
e \\
\end{bmatrix}

A line going through the origin is a valid vector space (vectors on the line).

$R^{3}$: all 3-dimensional real vectors, such as

\begin{bmatrix}
3 \\
2 \\
0 \\
\end{bmatrix}

\begin{bmatrix}
0 \\
0 \\
0 \\
\end{bmatrix}

\begin{bmatrix}
\pi \\
e \\
-1 \\
\end{bmatrix}

$R^{n}$: all column vectors with $n$ real components

##### Bad Example
In $R^{2}$ (the x-y plane), suppose the top right quadrant is a vector space, where vectors only have positive components. However, this cannot be a valid vector space since it's not "closed." For example, a vector can be multiplied by -1 to produce a vector outside the space.

#### Available Subspaces in $R^{n}$

##### All Subspaces of $R^{2}$
1) All of $R^{2}$
2) Any line through the origin
3) The zero vector only

##### All Subspaces of $R^{3}$
1) All of $R^{3}$
2) Any plane through the origin
3) Any line through the origin
3) The zero vector only
