### Video Lecture 4: Factorization into A = LU

#### A couple more inverse matrix facts

Starting with
$$
AA^{-1} = I
$$

Take the transpose of both sides.

$$
(A^{-1})^{T}A^{T} = I
$$

The transpose of $I$ is $I$ since its a symmetric matrix.

Note that the multiplication order of $A^{-1}$ and $A$ has been flipped. 

This is similar when in taking the inverse of the product of two matrices, the product of the inverses are flipped.

Somehow from the equation above, it is revealed that 

$$
(A^{-1})^{T} = (A^{T})^{-1}
$$

#### Number of Operations for Elimination

For an $n$ x $n$ matrix $A$, eimination takes $\frac{1}{3}n^3$ operations (counting multiply and subtract as a single operation).

For an $n$ x 1 column vector $b$, elimination takes $b^{2}$ operations.
