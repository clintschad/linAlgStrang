## Transposes and Permutations

### Transposes

#### Rules
* Sum: $(A+B)^{T} = A^{T} + B^{T}$
* Product: $(AB)^{T} = B^{T}A^{T}$ (note reversed order)
* Inverse: $(A^{-1})^{T} = (A^{T})^{-1}$

#### Inner Products
* Inner product (aka dot product) - $x^{T}y$ is (1 x n)(n x 1) - result is a number
* Outer product - $xy^{T}$ is (n x 1)(1 x n) - result is a matrix
* $(Ax)^{T}y = x^{T}(A^{T}y)$

#### Symmetric Matrices

A matrix is symmetric if $A^{T} = A$. This means $a_{ji} = a_{ij}$.

The inverse of a symmetric matrix is symmetric. Thus $(A^{-1})^{T} = (A^{T})^{-1} = A^{-1}$.

Given the matrix $R$, which may be square or rectangular, the product of it and its inverse, such that $R^{T}R$ or $RR^{T}$
- is square, with $R^{T}R$ being _n x n_ and $RR^{T}$ being _m x m_
- is symmetric
- diagonal is positive
- $R^{T}R$ does not necessarily equal $RR^{T}$

Very interesting note about factoring a symmetric matrix into $LDU$:
- if $A = A^{T}$ is factored into $LDU$ with no row exchanges, then $U$ is exactly $L^{T}$
- so $A = LDU = LDL^{T}$

This is great because it requires about half the work. The upper $U$ doesn't have to be found separately. It is simply the lower $L$ transposed.

#### Permutation Matrices

A permutation matrix $P$ has the rows of the identity matrix $I$ in any order.

There are $n!$ permutation matrices for an _n x n_ matrix.

Factoring with row exchanges: $PA = LU$
- $P$ is the product of all the permutuation matrices needed to swap the rows in $A$ to prepare it for the usual elimination
- Afterwards, perform elimination on $PA$ as usual
