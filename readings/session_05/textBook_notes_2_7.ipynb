## Transposes and Permutations

### Transposes

#### Rules
* Sum: $(A+B)^{T} = A^{T} + B^{T}$
* Product: $(AB)^{T} = B^{T}A^{T}$ (note reversed order)
* Inverse: $(A^{-1})^{T} = (A^{T})^{-1}$

#### Inner Products
* Inner product - $x^{T}y$ is (1 x n)(n x 1) - result is a number
* Outer product - $xy^{T}$ is (n x 1)(1 x n) - result is a matrix
* $(Ax)^{T}y = x^{T}(A^{T}y)$

#### Symmetric Matrices

A matrix is symmetric if $A^{T} = A$. This means $a_{ji} = a_{ij}$.

The inverse of a symmetric matrix is symmetric. Thus $(A^{-1})^{T} = (A^{T})^{-1} = A^{-1}$.
