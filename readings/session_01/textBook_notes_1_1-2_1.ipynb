### Section 1.2
#### Dot Products and Vectors
* The dot product (or _inner product_) of perpendicular vectors is 0 (scalar).
* Length of a vector $ \|v\| $ is $\sqrt{v \cdot v}$. Length squared is **v** $\cdot$ **v**. Note that $\|v\|$ is a scalar. 
* Unit vector $ \mathbf{u} = \frac{\textbf{v}}{\| \mathbf{v} \|}$, where $ \| \textbf{u} \| = 1$. Its length is always 1.
* The order of a dot product doesn't matter. $w \cdot v = v \cdot w $.

* In the _xy_ plane, **i** and **j** are unit vectors along the x and y axes.
* The unit vector that makes an angle _theta_ with the x-axis has components ($cos\theta, sin\theta)$.
* The angle between any two **unit** vectors is $\theta = arcos(\textbf{u} \cdot \textbf{U})$. Put another way, $cos(\theta) = \textbf{u} \cdot \textbf{U} $
* The dot product between any two unit vectors is always between -1 and 1.
* For non-unit (and non-zero) vectors, use the following to find the angle between them $cos\theta = \frac{\textbf{v} \cdot \textbf{w}}{\|v\|\|w\|}$
* The cosine between any 2 vectors never exceeds 1.
* Schwarz and Triangle inequalities mentioned in this section.

Dot product example in Python below

```python
import numpy as np

v = np.array([1, 2])
w = np.array([3, 4])

dot_product = np.dot(v, w)
print("Dot product of v and w is " + str(dot_product))
```
