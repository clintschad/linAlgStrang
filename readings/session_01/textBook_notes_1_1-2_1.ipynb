### Section 1.2
#### Dot Products and Vectors
* The dot product (or _inner product_) of perpendicular vectors is 0 (scalar).
* Length of a vector $ \|v\| $ is $\sqrt{v \cdot v}$. Length squared is **v** $\cdot$ **v**. Note that $\|v\|$ is a scalar. 
* Unit vector $ \mathbf{u} = \frac{\textbf{v}}{\| \mathbf{v} \|}$, where $ \| \textbf{u} \| = 1$. Its length is always 1.
* The order of a dot product doesn't matter. $w \cdot v = v \cdot w $.

* In the _xy_ plane, **i** and **j** are unit vectors along the x and y axes.
* The unit vector that makes an angle _theta_ with the x-axis has components ($cos\theta, sin\theta)$.
* The angle between any two **unit** vectors is $\theta = arcos(\textbf{u} \cdot \textbf{U})$. Put another way, $cos(\theta) = \textbf{u} \cdot \textbf{U} $
* The dot product between any two unit vectors is always between -1 and 1.
* For non-unit (and non-zero) vectors, use the following to find the angle between them $cos\theta = \frac{\textbf{v} \cdot \textbf{w}}{\|v\|\|w\|}$
* The cosine between any 2 vectors never exceeds 1.
* Schwarz and Triangle inequalities mentioned in this section.

Dot product example in Python below

```python
import numpy as np

v = np.array([1, 2])
w = np.array([3, 4])

dot_product = np.dot(v, w)
print("Dot product of v and w is " + str(dot_product))
```

### Section 1.3
#### Matrices

* Difference matrix. For $Ax = b$, elements in $b$ are the differences in $x$, going from top to bottom. So $b_{1} = x_{1}-0$, $b_{2} = x_{2}-x_{1}$, $b_{3} = x_{3}-x_{2}$, etc. 
    * Here, we're calling A the difference matrix. 
    * A is $\bf{invertible}$. 
    * All its columns are linearly independent.
    * No combination of the columns give the 0 vector, except for multiplying the columnns by 0. In other words, no Ax = 0, unless x = 0.
    * Therefore, Ax = 0 has only one solution, where x = 0.
    * The vectors of A fill the entire 3D space (since the example A is a 3x3).

* Cyclic difference matrix. Similar to difference matrix.
    * Here, we're calling C the cyclic difference matrix. 
    * C is $\bf{not invertible}$. C is $\bf{singular}$.
    * Its columns are $\bf{not}$ linearly independent. The last column is a sum of the other two.
    * Certain combinations of the columns give the 0 vector, such as the sum of the columns.
    * Cx = 0 has many solutions aside from x = 0.
    * The vectors of C are constrained to a 2D plane in 3D space, since it only has 2 independent vectors.

### Section 2.1
#### Vectors and Linear Equations

Different ways to view a matrix, or a system of equations put into a matrix, Ax = b
* Row picture - this is the familiar way we view systems of equations. If Ax = b, then row 1 of A is eq 1, row 2 of A is eq 2, etc. If b is given, the point b is where all the equations intersect. In a 2x2 matrix, or 2D, this is where the equation lines intersect. In a 3x3 matrix, or 3D, this is where the equations planes intersect. In higher dimensions, this is where the _hyperplanes_ intersect.
* Column picture - the columns of the matrix are combined to get the vector b. In other words, the column vectors are stretched, shortened, flipped, and added to get the b vector.

Identity matrix - matrix with diagonal filled with 1's and the rest are 0's. Multiplying against returns original matrix. This is analogous to multiplying a scalar by 1.

For Ax = b, to guarantee that specific x is the only x to produce that specific b, we must show that
* A is invertible
* The columns of A are independent
* and that the determinant is not 0
