## Readings: Sections 2.4 - 2.5

### Section 2.4
#### Rules for Matrix Operations

* For the matrix multiplication $AB$, if $A$ has $n$ columns, $B$ must have $n$ rows.
* If $A$ is $m \times n$ and $B$ is $n \times p$ then
$$
(m \times n )(n \times p) = (m \times p)
$$



$$

\begin{bmatrix}
\text{m rows}    \\
\text{n columns} \\
\end{bmatrix}

\begin{bmatrix}
\text{n rows}    \\
\text{p columns} \\
\end{bmatrix}

=

\begin{bmatrix}
\text{m rows}    \\
\text{p columns} \\
\end{bmatrix}

$$

* Row times column. Also called _inner product_ or _dot product_. Result is scalar or 1 $\times$ 1.

$$

\begin{bmatrix}
0 & 1 & 2 \\
\end{bmatrix}

\begin{bmatrix}
1 \\
2 \\
3 \\
\end{bmatrix}

=

8
$$

* Column times row. Also called _outer product_. If column is $n \times 1$ and row is $1 \times n$, then result is $n \times n$.

$$

\begin{bmatrix}
1 \\
2 \\
3 \\
\end{bmatrix}

\begin{bmatrix}
0 & 1 & 2 \\
\end{bmatrix}

=

\begin{bmatrix}
0 & 1 & 2 \\
0 & 2 & 4 \\
0 & 3 & 6 \\
\end{bmatrix}

$$

* For matrix multiplication $AB$, where $A$ and $B$ are matrices,
    * Column picture - each column of $B$ combines all columns of $A$ and stores into corresponding column of $AB$.
    * Row picture - each row of $A$ combines all rows of $B$ and stores into corresponding row of $AB$.
* With the exception of the commutative properties, most properties, such as distributive and associative, hold for matrix multiplication. Notable examples are $AI = IA$ and $A(cI) = (cI)A$ are commutative. 
* Also, exponents for matrices are the same as the scalar exponent operations we are familiar with. For example, $A^{p}A^{q} = A^{p+q}$ and $(A^{p})^{q} = A^{pq}$.

* Block Matrices
    * Adding by blocks allowed if dimensions of $A$ and $B$ are the same.
    * Multiplication is allowed if the "cuts" between columns of $A$ match the "cuts" between the rows of $B$.
    * Elimination also possible. See book for more detail.

### Section 2.5 Inverse Matrices

* $A^{-1}A = AA^{-1} = I$
* If there's a nonzero vector $x$ so that $Ax=0$, then $A^{-1}$ doesn't exist.
* For the inverse of a product, the order of multiplication is reversed: $(ABC)^{-1} = C^{-1}B^{-1}A^{-1}$
* Matrix is invertible if its determinant is $\textbf{not}$ zero.
* The product of the pivots of $A^{-1}$ is the determinant of $A$.
* $A^{-1}$ involves division by the determinant, which is why an invertible matrix cannot have a zero determinant. See page 85 of the text.
* Gauss-Jordan method can be used to find $A^{-1}$.

#### Triangular Matrices
* A triangular matrix is invertible iff no diagonal entries are zero.
* If $L$ is lower triangular with 1's on its diagonal, so is $L^{-1}$.
