## Subspaces

* A plane in $\mathbb{R}^{3}$ is not $\mathbb{R}^{2}$, even though it looks like $\mathbb{R}^{2}$. The vectors still have 3 components.
* A plane going through (0, 0,0 ) is a $\mathbf{subspace}$ of the full vector space $\mathbb{R}^{3}$.
* A subspace of a vector space is a set of vectors (including $\mathbf{0}$) that satisfies the following two requirements:
    - $v+w$ is in the subspace
    - $cv$ is in the subspace
* In other words, the subspace is _closed_ under addition and (scalar) multiplication. Furthermore, linear combinations of vectors in the subspace never produce a vector outside the subspace.

* All possible subspaces of $\mathbb{R}^3$:
    - Any line through (0, 0, 0)
    - Any plane through (0, 0, 0)
    - The whole space ($\mathbb{R}^{3}$)
    - The single vector (0, 0, 0)
* A subspace containing vectors $v$ and $w$ must contain all linear combinations $cv+dw$.
* The only "thing" (matrix, vector, function) that can be a subspace in and of itself is the 0 matrix, vector, or function, since every subspace must have this.

## Column Space

* Given $Ax = b$, if $A$ is not invertible, the system is solvable for certain $b's$ but not other $b's$.
* The $b's$ that can be produced from $Ax$ form the _column space_ of $A$.
* Use the column vectors of $A$ to produce every possible $b$ using every possible $x$.
* The column space consists of all the linear combinations of the columns of $A$.
* The column space of $A$ may also be referred to as $C(A)$.
* It can also be said that the column space of $A$ is _spanned_ by the columns.
* For $Ax = b$, $b$ must be in the column space of $A$, or there is no solution.
* Example: suppose $A$ is _3 x 2_, so two 3-dim vectors forming a plane subspace in $\mathbb{R}^{3}$. If $b$ is not in the plane, then there is no solution -- no $x$ or combo of the $A$ column vectors -- to produce the $b$ vector not in the plane.
* Since the columns of $A$ form the space, the space has the dimenionality of the column vectors. If $A$ is _m x n_, the column vectors are dimension _m_, so the column space of $A$ is $\mathbb{R}^{m}$.
