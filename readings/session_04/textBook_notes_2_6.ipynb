## Readings: Section 2.6

### Section 2.6
#### Elimination = Factorization: A = LU

##### The information for this section assumes no row exchanges.

$A$ can be factored into $LU$ when elimination matrices are applied to A

For $A = LU$,
* $U$ is the upper triangular matrix after doing row operations to find the pivots.
* $L$ is $E^{-1}$, where $E$ is the elimination matrix that does row operations on the left side of $A$ to turn $A$ into $U$, i.e. $EA = U$. Note that $E$ can consist of multiple matrices; however many row operations are needed to find the pivots of $A$.

$$
EA = U
$$

$$
A = E^{-1}U
$$

$$
A = LU
$$

* $E^-1$ is lower triangular.
* Lower triangular $L$ has 1's along its diagonal.
* For $LU$, $U$ may have values that are not 1 in its diagonal. In $LDU$, $U$ has 1's in the diagonal. To get $LDU$ from $LU$, divide $U$ by a diagonal matrix $D$ that contains the pivots.

Algorithm to solve $Ax = b$ using $A = LU$.
* Look at Example 3 on page 98. Essentially breaks the problem into two matrix equations: first solve $Lc = b$ and then solve $Ux = c$, where $c$ is the right hand side produced during forward elimination (Gauss Jordan method). It appears when solving $Lc = b$ and $Ux = c$, Gauss Jordan is still being used. 
* However, something I noticed is for $Lc = b$, you could use $E^{-1}Lc = E^{-1}b$ which equals $c = E^{-1}b$ to find $c$ through direct multiplication without having to use Gauss Jordan.

Computation cost of elimination
* Elimination on $A$ requires about $\frac{1}{3}n^3$ multiplications and $\frac{1}{3}n^3$ subtractions.
* Elimination on $b$ requires $n^2$ multiplications and $n^2$ subtractions.
