### Section 2.4
#### Rules for Matrix Operations

* For the matrix multiplication $AB$, if $A$ has $n$ columns, $B$ must have $n$ rows.
* If $A$ is $m \times n$ and $B$ is $n \times p$ then
$$
(m \times n )(n \times p) = (m \times p)
$$



$$

\begin{bmatrix}
\text{m rows}    \\
\text{n columns} \\
\end{bmatrix}

\begin{bmatrix}
\text{n rows}    \\
\text{p columns} \\
\end{bmatrix}

=

\begin{bmatrix}
\text{m rows}    \\
\text{p columns} \\
\end{bmatrix}

$$
