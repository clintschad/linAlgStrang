## Readings: Sections 2.2 - 2.3

### Section 2.2
#### The Idea of Elimination
* $\textbf{Elimination}$ produces upper triangular system. This just means elements below the diagonal are zero. The diagonal doesn't necessarily need to be 1's.
* After elimination, can use back substitution to find the rest of the variables.
* $\textbf{Pivot}$ - diagonal element, which is a variable coefficient. The elements to its left, if present, must be zero. For example, the first element of a square matrix is the first pivot. In elimination, the goal is to get these to equal 1.
* A zero pivot point indicates that row, or equation line/plane, is parallel to another line/equation. Since they're parallel, they never intersect, so there's no value they share, or that satisfies both of them. This is the row picture.
* A zero pivot, from a column perspective, indicates the two vectors are on the same line, and will thus never produce a vector that's not pointed, or in line, with that line.
* $\textbf{Nonsingular}$ - 1 pivots along the entire diagonal. These have a solution to the system of equations.
* $\textbf{Singular}$ - a zero pivot in the diagonal. These have either no solution or infinitely many.
* From a row perspective, for a nonsingular 3x3 matrix, the equations are 3 planes with various slopes. When you complete forward elimination, the plane for the last equation gives you a flat, horizontal slope at a certain z height. Continuing back substitution will make vertical planes slicing x and y axes. In other words, the original 3 sloped planes intersected at a point. Doing full elimination keeps the planes intersecting at that point, but angles the planes perpendicular to their respective axes. It's like elimination is "straightening" out the planes in a way.

### Section 2.3

#### Elimination Using Matrices

* Row picture of a matrix - for a 2x2 matrix, at what (x,y) point in 2D space do the (row/equation) lines intersect? For a 3x3 matrix, at what (x,y,z) point in 3D space do the (row/equation) planes intersect? For an nxn matrix, at what (x,y,..n) point in nD space do the hyperplanes intersect?
* Column picture of a matrix - for Ax=b, what combination (stretch, flip, add) of the column vectors (of A) are needed to create the vector b?
* "Entry" for matrix corresponds to "component" for vector.
* Matrix multiplication is associative: A(BC) = (AB)C
* Matrix multiplication is NOT commutative: AB != BA

* Assume E is a modified 3x3 identity matrix, with -2 in the (2,1) position.
    * EA subtracts 2 times row 1 from row 2. Multiplying on the left does row operations.
    * AE subtracts 2 times col 1 from col 2. Multiplying on the right does col operations.

* The matrix above can also be called $E_{21}$, where it take a multiple $l_{21}$ of row 1 and adds it to row 2 (result placed in row 2). The number $l_{21}$ is the (2,1) entry.

* Multiplying a matrix by another matrix can be seen as multiplying a matrix by each column of the other matrix. In above example, EA can be seen as E x A_col1, E x A_col2, and E x A_col2.

* Row exchange / permutation matrix - a matrix that exchanges rows of the vector / matrix that it's being multiplied against. To create a row exchange matrix, take an identity matrix and exchange the desired rows.

* Below is an example of a column times a vector:

$$

\begin{bmatrix}
3 \\
1 \\
2 \\
\end{bmatrix}
\begin{bmatrix}
2 & 4 \\
\end{bmatrix}
=
\begin{bmatrix}
6 & 12 \\
2 &  4 \\
4 &  8 \\
\end{bmatrix}

$$

Viewing the vector operation above, multiplying right to left is the column picture, and multiplying left to right is the row picture.
