Element of matrix (row #, col#)

When 2 matrices are multiplied, the element (3, 4) in the resulting product matrix comes from row 3 of the matrix on the left multiplied by column 4 of the matrix on the right.

In other words, for A*B = C, where A, B, and C are matrices,

$C_{3,4}$ = (row 3 of A) * (col 4 of B)

$$
= a_{3,1}b_{1,4} + a_{3,2}b_{2,4} + ...

= \sum_{k=-1}^{n} a_{3k}b_{k4}
$$
