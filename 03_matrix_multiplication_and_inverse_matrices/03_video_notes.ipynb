Element of matrix (row #, col#)

When 2 matrices are multiplied, the element (3, 4) in the resulting product matrix comes from row 3 of the matrix on the left multiplied by column 4 of the matrix on the right. This is also called the "dot product" of the row and column.

In other words, for AB = C, where A, B, and C are matrices,

$C_{3,4}$ = (row 3 of A) * (col 4 of B)

$$
= a_{3,1}b_{1,4} + a_{3,2}b_{2,4} + ...

= \sum_{k=-1}^{n} a_{3k}b_{k4}
$$

### Valid dimensions for multiplying

A is "m x n" or "m rows by n columns."

If AB, and A is mxn, then B must be nxp, where p can be any number. In other words, the "height" of B must equal the "width" of A.

In the above example, let's say AB = C, where A is mxn, B is nxp. Then C is mxp.

### Other ways of thinking about matrix multiplication

For AB = C, one can think of A being multiplied against each column of B separately, such that A x B_col1 = C_col1, and A x B_col2 = C_col2, .... Then C is just all the resulting columns together. In other words, each column in C is a combo of A columns.

Similarly, each row in C is a combo of B rows. The row in A defines how many of each row in B to add up, resulting in the corresponding row in C.

### Summary of matrix multiplication methods
1) Regular method - row of A x col B = single entry in the matrix
2) Row method - resulting matrix is a combination of rows
3) Column method - resulting matrix is a combination of columns
4) Col x row - col A x row B = matrix, since mx1 times 1xp = mxp. How GS puts it, AB = sum of (columns of A) times (rows of B).
5) Block method - you can actually "cut" the matrices into sections and perform method 1 (regular way). See image.

## Inverse Matrices

For all matrices, if $A^{-1}$ exists (aka matrix is _invertible_ or _nonsingular_), then A times its inverse equals the identity matrix.
$$
A^{-1}A = I
$$

For square matrices, multiplying by the inverse on either side works.
$$
A^{-1}A = AA^{-1} = I
$$

This does not hold for rectangular matrices.

### Singular case (no inverse)

The determinant of a singular matrix is 0.

If a (non-zero) vector can be found, such that Ax = 0, then the matrix is not-invertible.

Simple "proof". See image.

Suppose you have found the non-zero vector such that Ax = 0. If you multiply by the inverse on both sides, $A^{-1}Ax = A^{-1}0$, then $Ix = 0$ or $x=0$, but we said we started with a non-zero vector.

## Back to invertible matrices
A x column of A inverse = column of I

Gauss-Jordan method example. Basically, augment identity matrix on RHS of A and use elimination to get identity matrix on LHS. Whatever the result on the RHS is A inverse.
