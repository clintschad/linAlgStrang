Pivot - the number in the matrix row you're trying to eliminate. For example, in a 3x3 system, row one column one would be the first pivot since that's the first variable we're trying to eliminate.

The pivots are the elements of the diagonal.

Upper triangular means elements below the diagonal are 0. Sometimes these matrices are called "U".

Can easily find determinant of upper triangular matrix by multiplying the pivots (elements of the diagonal).

If not possible to avoid zero in a pivot position, matrix is not invertible.

Augmented matrix - when RHS b is an added column.

Matrix x column = column (operation is combining columns)

row x matrix = row  (operation is combining rows)

"Elementary" matrix or "elimination" matrix - matrix in mid process during elimination.

"Permutation" matrix. Looks like an identity matrix with the diagonal going the other way. 
- Multiplying a matrix with a permutation matrix on the right side exchanges columns of the other matrix. 
- Multiplying a matrix with a permutation matrix on the left side exchanges rows of the other matrix. 

In general, multiplying a matrix with a matrix on the left does row operations. Multiplying a matrix with a matrix on the right does column operations.

Associative law (moving the parentheses in matrices being multiplied) holds. The commutative law (swapping order of matrices being multiplied) does **not** hold.
