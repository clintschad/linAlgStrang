### Calculus

#### Integration

Indefinite Integral
$$
\int x^2 \, dx = \frac{1}{3}x^{3} + C
$$

Definite Integral
$$
\int_{-\infty}^{\infty} \frac{dx}{dt}y(t)dt = \int_{-\infty}^{\infty}x(t)\left(-\frac{dy}{dt}\right)dt
$$
