#### Identity Matrix

#### Inverse Matrix

```python
import numpy as np

A = np.array([[1,2,3],
              [2,5,7],
              [3,8,9]])
A_inv = np.linalg.inv(A)

print("A inverse: ")
print(A_inv)
```
